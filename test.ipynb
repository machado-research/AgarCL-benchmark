```python
import numpy as np 

import matplotlib.pyplot as plt

# Load the file
file_name = '/home/mamm/ayman/thesis/AgarLE-benchmark/results/32a1d917a781803b591936a99be084fab60a6381-3481cf7c-d93b034b/episode_rewards.txt'
with open(file_name, 'r') as file:
    data = file.read()

# Parse the data
lines = data.strip().split('\n')
x = []
y = []
for line in lines:
    key, value = line.split(':')
    x.append(int(key.strip()))
    y.append(float(value.strip()))


# Plot the data
plt.plot(x, y, marker='o')
plt.xlabel('Number of Steps')
plt.ylabel('Reward')
plt.title('Number of Steps vs Reward')
plt.show()
```

```python
import imageio
import os

# Define the directory containing the images
image_folder = '/home/mamm/ayman/thesis/AgarLE-benchmark/results/32a1d917a781803b591936a99be084fab60a6381-75e33b91-d93b034b/images'
video_name = 'output_video.mp4'

# Get the list of images
images = [f"{i}.png" for i in range(12846, 35349)]

# Create a video writer object
with imageio.get_writer(video_name, fps=20) as writer:
    for image in images:
        image_path = os.path.join(image_folder, image)
        writer.append_data(imageio.imread(image_path))
```
